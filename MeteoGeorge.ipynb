```python
from tkinter import *

import requests



window=Tk()
window.title("Meteorology")
window.geometry('450x300')
window.resizable(width=False,height=False)
window.configure(background="beige")

def search():
    city=city_entry.get()
    
    API = "https://api.openweathermap.org/data/2.5/weather?q={}&lang=en&units=metric&appid=959b696984280a9e3dcb2c071fd76f4d".format(city)
    jsonData = requests.get(API).json()
    description = jsonData['weather'][0]['description']
    
    temperature = jsonData['main']['temp']
    tempo=temperature,u'\N{DEGREE SIGN}'"C"
    
    humidity = jsonData['main']['humidity']
    humid=humidity,"%"
    
    pressure=jsonData['main']['pressure']
    press=pressure,"hPa"
    
    descr_out.delete(0.0,END)
    descr_out.insert(END,description)
    descr_out.tag_configure("center",justify="center")
    descr_out.tag_add("center", 1.0, "end")

    
    
    Temp_out.delete(0.0,END)
    Temp_out.insert(END,tempo)
    
    
    Hum_out.delete(0.0,END)
    Hum_out.insert(END,humid)
    
    
    Press_out.delete(0.0,END)
    Press_out.insert(END,press)
    


    
Temp_lbl=Label(window,text="Temperature",bg="purple",fg="white")
Temp_lbl.grid(column=0,row=6,padx=20,pady=15)
    
Temp_out=Text(window,width=8,height=1)
Temp_out.place(x=120,y=110)
    
    
Hum_lbl=Label(window,text="Hummidity",bg="blue",fg="white")
Hum_lbl.grid(column=0,row=7,padx=20,pady=15)
    
Hum_out=Text(window,width=7,height=1)
Hum_out.place(x=120,y=160)
    
    
    
Press_lbl=Label(window,text="Pressure",bg="orange",fg="white")
Press_lbl.grid(column=0,row=8,padx=20,pady=15)
    
Press_out=Text(window,width=8,height=1)
Press_out.place(x=120,y=210)


city_entry=Entry(window,width=15)
city_entry.grid(column=0,row=2)


    
    
descr_lbl=Label(window,text="Weather Description",bg="gray",fg="white")
descr_lbl.grid(column=4,row=5,padx=130)
    
    
descr_out=Text(window,width=20,height=1)
descr_out.place(x=220,y=125)

    
search_btn=Button(window,text="Search City",bg="green",fg="white",command=search)
search_btn.grid(column=0,row=0,padx=20,pady=15)




    
window.mainloop()
```
